Теперь выполним начальную алгебраическую выставку на заранее подобранном неплохом кусочке (в прошлой версии он каждый раз пересчитывал довольно тяжёлый кусок кода для поиска интервала минимального вибрационного воздействия, я от этого отказался)

```python
def InitialAlgebraicAlignment():
    global g
    global L
    global theta, gamma, phi, psi, psi_al, theta_al, gamma_al, L_al
    DELIMITER0 = 29986  #начало интервала наименьшего вибрационного воздействия
    DELIMITER = 50004   #конец периода наименьшего вибрационного воздействия (получены на первой курсовой)
    u = 0.000072921158553
    Pi = 3.14159265358979323846
    om_x = CNFnav['OmX'][DELIMITER0:DELIMITER].mean()
    om_y = CNFnav['OmY'][DELIMITER0:DELIMITER].mean()  
    om_z = CNFnav['OmZ'][DELIMITER0:DELIMITER].mean()
    f_x = CNFnav['AccX'][DELIMITER0:DELIMITER].mean()
    f_y = CNFnav['AccY'][DELIMITER0:DELIMITER].mean()
    f_z = CNFnav['AccZ'][DELIMITER0:DELIMITER].mean()
    theta = math.atan2(f_x, math.sqrt(f_y*f_y+f_z*f_z))
    gamma = -math.atan2(f_z, f_y)
    phi = 49.99453181*math.pi/180
    g = 9.78030*(1+0.005302*sin(phi)*sin(phi)-0.000007*sin(2*phi)*sin(2*phi))
    om = math.sqrt(om_x*om_x+om_y*om_y+om_z*om_z)
    l13 = sin(theta)
    l23 = cos(theta)*cos(gamma)
    l33 = -cos(theta)*sin(gamma)
    l12 = (om_x - u*l13*sin(phi))/(u*cos(phi))
    l22 = (om_y - u*l23*sin(phi))/(u*cos(phi))
    l32 = (om_z - u*l33*sin(phi))/(u*cos(phi))
    l12 = l12/(sqrt(l12*l12+l22*l22+l32*l32)) #нормировка
    l22 = l22/(sqrt(l12*l12+l22*l22+l32*l32))
    l32 = l32/(sqrt(l12*l12+l22*l22+l32*l32))
    #теперь нужно векторно умножить второй на третий
    l11 = l22*l33 - l32*l23
    l21 =-l12*l33 + l13*l32
    l31 = l12*l23 - l22*l13
    L = np.matrix([[l11, l12, l13], [l21, l22, l23], [l31, l32, l33]])   #записали в матрицу
    psi = math.atan2(L.item(0,0),L.item(0,1))
    theta = math.asin(L.item(0,2))
    gamma = -math.atan2(L.item(2,2),L.item(1,2))
    psi_al = psi                                      
    theta_al = theta
    gamma_al = gamma
    L_al = L
```

# Фильтр Калмана
Займёмся дизайном фильтра - сформулируем мат. модель и построим начальное состояние фильтра. Далее в цикле пройдём до конца интервала выставки и определим ошибки алгебраической выставки

```python
def StateUpdate(i): #эта функция формирует переходную матрицу непрерывного процесса. Позднее её дискретизируем
    global A
    A[0,1] = CNFnav['OmZ'][i]
    A[0,2] = -CNFnav['OmY'][i]
    A[1,0] = -A[0,1]
    A[1,2] = CNFnav['OmX'][i]
    A[2,0] = -A[0,2]
    A[2,1] = -A[1,2]
    A[0,6] = 1
    A[1,7] = 1
    A[2,8] = 1

def MeasurmentMatrixUpdate(): #обновляем матрицу измерений
    global H                    #на выставке нужно отслеживать phi и пересчитывать по формуле Гелмерта иногда
    global theta, gamma
    gy1 = -g*sin(theta)
    gy2 = -g*cos(theta)*cos(gamma)
    gy3 = -g*cos(theta)*sin(gamma)
    H[0,1] = gy3
    H[0,2] = -gy2
    H[0,3] = 1
    H[1,0] = -gy3
    H[1,2] = gy1
    H[1,4] = 1
    H[2,0] = gy2
    H[2,1] = -gy1
    H[2,5] = 1

def LUpdate():          #функция обновления матрицы поворота относительно географической системы координат
    global theta, gamma, psi, phi, L, KF
    theta = (theta - (-KF.x.item(0)*cos(psi) + KF.x.item(1)*sin(psi))) 
    gamma = (gamma - (-KF.x.item(0)*sin(psi) + KF.x.item(1)*cos(psi))/cos(theta))
    psi = (psi - KF.x.item(2) - gamma*sin(theta))
    
    l11 = sin(psi)*cos(theta)
    l21 = cos(psi)*sin(gamma) - sin(psi)*sin(theta)*cos(gamma)
    l31 = cos(psi)*cos(gamma) + sin(psi)*sin(theta)*sin(gamma)

    l12 = cos(psi)*cos(theta)
    l22 = -sin(psi)*sin(gamma) - cos(psi)*sin(theta)*cos(gamma)
    l32 = -sin(psi)*cos(gamma) + cos(psi)*sin(theta)*sin(gamma)

    l13 = sin(theta)
    l23 = cos(theta)*cos(gamma)
    l33 = - cos(theta)*sin(gamma)
    
def DegreesUpdate():      #функция обновляющая значения малых углов поворота
    global theta, gamma, psi, phi, L, KF
    theta = (theta - (-KF.x.item(0)*cos(psi) + KF.x.item(1)*sin(psi))) 
    gamma = (gamma - (-KF.x.item(0)*sin(psi) + KF.x.item(1)*cos(psi))/cos(theta))
    psi = (psi - KF.x.item(2) - gamma*sin(theta))
```

Определили функции, запустим фильтр:

```python
b1, b2, b3, xs, ps = [], [], [], [], []
CNFnav = CNFnav[:][0:78537]            #оставили данные только для начальной выставки - первые 78537 
                                       #измерений до начала движения
L = L_al                               #инициализировали матрицу поворота и углы ориентации дефектоскопа по
psi = psi_al                           #результатам начальной выставки
theta = theta_al
gamma = gamma_al
KF = KalmanFilter(dim_x=9, dim_z=3)  #создали объект класса: фильтр Калмана методом квадратного корня
KF.x = np.array([[0],[0],[0],[6.7e-5],[-5e-3],[-3.9e-3],[8.4e-8],[-9.8e-7],[-9e-7]]) #начальное положение
A = np.zeros((9,9))                     #инициализировали A нулями явно, на всякий случай
StateUpdate(0) #возвращает глобальную матрицу А непрер. процесса: x' = Ax + Bu + Gq где G интенсивность (scale) шума
dt = 0.01      #нужно бы считать честно: print CNFnav[0]['Time'] - CNFnav[1]['Time'] 
G = np.array([[0.01],[0.01],[0.01],[1e-3],[1e-3],[1e-3],[1e-5],[1e-5],[1e-5]]) # scale шума с масштабом: на малых углах 0,
#для измерений силы тяжести 1, на ДУС примерно dt
F1, Q = van_loan_discretization(A, G, dt)  #дискретизировали A (т.е. получили F) и сразу красивая дискретная Q
                                        #про процедуру Ван Лоана есть в его книжке
KF.Q = Q                         #неопределённость процесса по сути
KF.F = F1                                  # Дискретизированная матрица переходного процесса
H = np.zeros((3,9))                        #инициализировали H нулями явно (z = Hx + шум)
MeasurmentMatrixUpdate()                   #вызвали процедуру обновления матрицы измерений
KF.H = H                                   # Measurement function
KF.P *= 10                               #по-хорошему в матрицу ковариаций нужно записать начальные ковариации,
                                           #но я считаю начальное приближение довольно неточным (нач. ков. порядка 1e-6)
KF.predict()
for i, Measurment in enumerate(CNFnav):
    StateUpdate(i)
    z = np.transpose(np.dot(np.transpose(L),np.array([CNFnav[i]['AccX'],CNFnav[i]['AccY'] , CNFnav[i]['AccZ']])) - np.array([0,0,g]))
    KF.update(z)                             #строчкой выше - формируем корректирующее измерение
    KF.predict()
    MeasurmentMatrixUpdate()
    KF.H = H
    F1, Q = van_loan_discretization(A, G, dt) 
    KF.F = F1
    b1.append(KF.x.item(0))
    b2.append(KF.x.item(1))
    b3.append(KF.x.item(2))
    xs.append(KF.x)
    ps.append(KF.P.diagonal())   #попутно сохраняю ковариации (они лежат на диагонали)
LUpdate() #в конце выставки корректируем углы
beta = np.asarray(xs)
p = np.asarray(ps)
```

Продолжим считать дальше: закодим формулы из методички INSdemo и построим траекторию движения дефектоскопа

```python
def smooth(x,window_len=11,window='hanning'):
    if x.ndim != 1:
        raise ValueError, "smooth only accepts 1 dimension arrays."
    if x.size < window_len:
        raise ValueError, "Input vector needs to be bigger than window size."
    if window_len<3:
        return x
    if not window in ['flat', 'hanning', 'hamming', 'bartlett', 'blackman']:
        raise ValueError, "Window is on of 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'"
    s=np.r_[x[window_len-1:0:-1],x,x[-1:-window_len:-1]]
    #print(len(s))
    if window == 'flat': #moving average
        w=np.ones(window_len,'d')
    else:
        w=eval('numpy.'+window+'(window_len)')
    y=np.convolve(w/w.sum(),s,mode='valid')
    return y
```
